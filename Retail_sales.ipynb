## Data Cleaning

### Formating Column names

```python
df.columns=[col.title() for col in df.columns]
df.columns = df.columns.str.replace('_', ' ')
df
```

```python
df = df.rename(columns={'Cogs' : 'Cost of Goods Sold'})
df = df.rename(columns={'Quantiy' : 'Quantity'})
df
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

### Cleaning Age Column

```python
# Identifying null values
df[df['Age'].isnull()]
```

### Checking for Distribution of Values

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(df['Age'],kde=True)
```

### Filling NaN values with median as the chart shows the uniform distribution with no distinct peak. 

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df
```

```python
df[df['Age'].isnull()]
```

```python
df.isnull().sum()
```

## Cleaning Quantity, Price_Per_Unit, Cost of Goods Sold and Total_Sales Column

```python
df[df['Quantity'].isnull()] 
```

#### The columns "Quantity," "Price Per Unit," "Cost of Goods Sold," and "Total Sale" contain null values in the same rows, indicating that these may represent failed or incomplete transactions. Given that only three such records exist, it is reasonable to remove them from the dataset to maintain data quality.

```python
df.dropna(subset=['Quantity','Price Per Unit', 'Cost of Goods Sold','Total Sale'],inplace=True)
```

```python
df.isnull().sum()
```

#### All missing values have been handled, and the dataset is now clean and ready for analysis.

## Exporting the cleaned dataset

```python
df.to_csv('Retail sales cleaned.csv', index=False)
```
